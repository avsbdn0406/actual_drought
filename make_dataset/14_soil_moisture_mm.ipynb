## 2020

```python
files = glob.glob(year_list[-2]+'/*')
files.sort()
files
```

```python
data = pd.read_excel(files[0],header=3,sheet_name=None)
sheet_names = [k for k in data.keys()]
sheet_names
```

```python
year = '2020'
columns_list = ['Region','Point','Address', 'Depth(cm)','Year','Month','Day','Moisture(%)']
```

```python
for f in files:
    # Make the list of sheet names
    res = pd.DataFrame(columns=columns_list)
    
    
    data = pd.read_excel(f,header=3,sheet_name=None)
    sheet_names = [k for k in data.keys()]

    for s in sheet_names:
        data = pd.read_excel(f,header=1,sheet_name = s)
        addr = data['Unnamed: 12'][0].split(',')[0].split(':')[1][1:]
        print(addr)

        # Get the region name
        data = pd.read_excel(f, sheet_name = s)
        region = data.keys()[0].split(' ')[0]
        # Get the dataframe
        data = pd.read_excel(f, header = 3, sheet_name = s)
        # Except for the month (ENG)
        data = data.drop([0])
        # Except the outliers columns
        data = data[data.columns[:-2]]
        # Just get the moisture info.
        data = data[:31].reset_index(drop=True)

        # Get the point and depth info.
        point = s.split('-')[0]
        depth = s.split('-')[-1]
        print(region, point, depth)

        
        for m in data.columns[1:]:
            for d in data[data.columns[0]]:
                # If data is not NONE
                if data[m][d-1]!= ' ':
                    items = pd.DataFrame({'Region':region,
                                          'Point' : point,
                                          'Address' : addr,
                                          'Depth(cm)' : depth,
                                          'Year':year, 
                                          'Month' :m, 
                                          'Day' : str(d),
                                          'Moisture(%)' : data[m][d-1]}, index=[0])

                    res = pd.concat([res,items],ignore_index=True)
                else:
                    pass

        res = res.dropna(axis=0)
        # res = res.fillna(0)
    res.to_csv('../results_data/14_토양수분량_'+region+'_'+year+'.csv',index=False,encoding='cp949')
```
