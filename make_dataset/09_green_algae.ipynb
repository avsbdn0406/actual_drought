```python
import os
import glob
import pandas as pd
import shutil
from itertools import repeat
import itertools
```

```python
add_file = '../샘플데이터/09. 물환경정보시스템 - 녹조(조류)정보/녹조(조류)정보주소.xlsx'
data = pd.read_excel(add_file,sheet_name=None)
sheet_names = [k for k in data.keys()]
sheet_names
```

```python
# Get the address
addr = pd.DataFrame()
for s in sheet_names:
    df = pd.read_excel(add_file,sheet_name=s)
    df['데이터구분'] = list(repeat(s,len(df)))
    addr = pd.concat([addr,df])
addr = addr.reset_index(drop=True)
```

```python
# Load the dataframe
file_01 = '../샘플데이터/09. 물환경정보시스템 - 녹조(조류)정보/상수원구간일자료_2000-2022.csv'
file_02 = '../샘플데이터/09. 물환경정보시스템 - 녹조(조류)정보/조류관찰지점일자료_2012-2022.csv'
file_03 = '../샘플데이터/09. 물환경정보시스템 - 녹조(조류)정보/친수활동구간일자료_2007-2022.csv'
df_01 = pd.read_csv(file_01,encoding='cp949')
df_01['데이터구분'] = list(repeat(sheet_names[0],len(df_01)))
df_02 = pd.read_csv(file_02, encoding='cp949')
df_02['데이터구분'] = list(repeat(sheet_names[1],len(df_02)))
df_03 = pd.read_csv(file_03, encoding='cp949')
df_03['데이터구분'] = list(repeat(sheet_names[2],len(df_03)))
                      
```

```python
df = pd.concat([df_01,df_02,df_03]).reset_index(drop=True)
```

```python
res_addr = []
for d in df.iloc:
    region = d['지점명']
    loc = d['채수위치']
    cate = d['데이터구분']
    res_addr.append(addr[(addr['지점명']==region) & (addr['조사지점']==loc)]['주소'].iloc[0])
```

```python
df['주소'] = res_addr
```

```python
c = df.keys().to_list()
cols = [c[0],c[1],c[2],c[-1],c[-2]]
cols.extend(c[3:-2])
```

```python
df = df.reindex(cols,axis=1)
df
```

```python
df.to_csv('../results_data/09_녹조조류정보_2000_2022.csv',index=False,encoding='cp949')


```
