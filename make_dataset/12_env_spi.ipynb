```python
import os
import glob
import pandas as pd
import shutil
from itertools import repeat
import itertools
```

```python
files = glob.glob('../샘플데이터/12. 기상자료개방포털_가뭄지수(SPI)/2021/*')
```

```python
regions = ['속초','철원', '태백']
```

```python
res = pd.DataFrame()
for i, f in enumerate(files[:3]):
    df = pd.read_csv(f,encoding='cp949')
    df['지점명'] = list(repeat(regions[i],len(df)))
    res = pd.concat([res,df])
                         
```

```python
c = res.keys().to_list()
cols = [c[0],c[-1]]
cols.extend(c[1:-1])
```

```python
res = res.reindex(cols,axis=1)
res
```

```python
res.to_csv('../results_data/12_가뭄지수_2021.csv',encoding='cp949',index=False)
```
