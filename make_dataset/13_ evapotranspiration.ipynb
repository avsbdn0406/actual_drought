```python
files = glob.glob(year_list[5]+'/*.xls')
```

```python
data = pd.read_excel(files[0],header=3,sheet_name=None)
sheet_names = [k for k in data.keys()]
```

```python
sheet_names
```

```python
data = pd.read_excel(files[0],header=3,sheet_name = sheet_names[0])
data = data.drop([0])
data = data.reset_index(drop=True)
data = data[:31]
```

```python
year = '2015'
columns_list = ['Region','Address', 'Year','Month','Day','Evapotrans(mm/d)','GapFilled']
```

```python
res = pd.DataFrame(columns=columns_list)
for region in sheet_names:
    data = pd.read_excel(files[0],header=1,sheet_name = region)
    addr = data['Unnamed: 12'][0].split(',')[0].split(':')[1][1:]
    print(addr)
    
    data = pd.read_excel(files[0],header=3, sheet_name = region)
    data = data.drop([0])
    data = data.reset_index(drop=True)
    data = data[:31]
    region = re.sub(r'[0-9]','', region)

    for m in data.columns[1:13]:
        for d in data[data.columns[0]]:
            if str(data[m].iloc[d-1])[-1] == '*':
                items = pd.DataFrame({'Region':region,
                                      'Address': addr,
                                     'Year':year, 
                                     'Month' :m, 
                                     'Day' : str(d),
                                     'Evapotrans(mm/d)' : str(data[m][d-1])[:-1],
                                     'GapFilled' : '1'}, index=[0])
            else:
                if str(data[m].iloc[d-1]) == ' ':
                    break
                else:
                    items = pd.DataFrame({'Region':region,
                                          'Address':addr,
                             'Year':year, 
                             'Month' :m, 
                             'Day' : str(d),
                            'Evapotrans(mm/d)':data[m][d-1],
                             'GapFilled' : '0'}, index=[0])
            res = pd.concat([res,items],ignore_index=True)
            
    res = res.dropna(axis=0)
res.to_csv('../results_data/13_일증발산량_'+year+'.csv',index=False,encoding='cp949')
```
