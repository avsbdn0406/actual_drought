```python
regions = pd.read_excel('../샘플데이터/10. 물환경정보시스템_수질측정망/지역코드명.xlsx',header=1)
```

```python
regions = regions[['지점명','주소']]
```

```python
df = pd.read_excel('../샘플데이터/10. 물환경정보시스템_수질측정망/전국수질측정자료_2022년_일자료.xlsx')
df_02 = pd.read_csv('../샘플데이터/10. 물환경정보시스템_수질측정망/자료 조회_20230210_수질측정망_하천.csv',encoding='cp949')
```

```python
df = df[df.keys()[1:]]
```

```python
df_list = df.keys().to_list()
df_list.sort()
for i, d in enumerate(df_list):
    if ',' in d:
        print(d)
        df = df.rename(columns={d:d.replace(',','.')})
    elif ' ' in d:
        df = df.rename(columns={d:d.replace(' ', '')})

org_list = df_02.keys().to_list()
org_list.sort()
```

```python
new = []
for d in df_list:
    for o in org_list:
        if d[0]=='1':
            if d[:3] == o[:3]:
                print(d, o)
                new.append(o)
                df = df.rename(columns={d:o})
        elif d[:2] == o[:2]:
            print(d, o)
            new.append(o)
            df = df.rename(columns={d:o})
```

```python
df.keys()
```

```python
add_list = []
for d in df.iloc:
    if len(regions[regions['지점명'] == d['측정소명']])!=0:
        add_list.append(regions[regions['지점명'] == d['측정소명']]['주소'].iloc[0])
    else:
        add_list.append('-')
    
```

```python
df['지점주소'] = add_list
```
